## Upper coil: Red -> '+' (power supply coonection), Black -> '-'; Lower coil: Red -> '-', Black -> '+'

```python
import matplotlib.pyplot as plt
import numpy as np

# Given data, z_28 data is for z-values at current 2.8A
z_values = {
    'z_28': [0, 2, 4, 6, 8, 10, 12, 13.68, 16, 18, 20, 22, 24, 25.5],
    'z_32': [0, 2, 4, 6, 8, 10, 12, 13.60, 16, 18, 20, 22, 24, 25.5],
    'z_36': [0, 2, 4, 6, 8, 10, 12, 13.66, 16, 18, 20, 22, 24, 25.5],
    'z_40': [0, 2, 4, 6, 8, 10, 12, 13.62, 16, 18, 20, 22, 24, 25.5],
    'z_44': [0, 2, 4, 6, 8, 10, 12, 13.67, 16, 18, 20, 22, 24, 25.5]
}

B_values = {
    'B_28': [-37.7, -32.8, -27.6, -22.2, -16.6, -10.8, -4.9, 0, 6.6, 12.3, 18, 23.5, 28.8, 32.6],
    'B_32': [-42.9, -37.2, -31.2, -25.0, -18.6, -12, -5.3, 0, 7.8, 14.5, 20.9, 27.2, 33.3, 37.6],
    'B_36': [-48.1, -41.2, -35.1, -28.2, -20.9, -13.6, -6.2, 0, 8.8, 16.3, 23.6, 30.6, 37.5, 42.4],
    'B_40': [-53.2, -46.2, -38.8, -31.1, -23, -14.9, -6.6, 0, 10.2, 18.4, 26.6, 34.5, 42, 48],
    'B_44': [-58.0, -50.5, -42.6, -34.4, -25.8, -16.8, -7.7, 0, 11.8, 21, 29.9, 38.7, 46.9, 52.9]
}

# Plotting
fig, ax = plt.subplots(1, 1, figsize=(10, 6), dpi=120)

# Use a for loop to plot each z value set against its corresponding B value set
for (z_label, z), (B_label, B) in zip(z_values.items(), B_values.items()):
    ax.plot(z, B, 'o', label=f'{z_label} vs {B_label}')
    
    # Perform linear fit
    m, c = np.polyfit(z, B, 1)  # Linear fit (degree 1)
    fit_line = np.polyval([m, c], z)  # Compute fitted line
    
    # Plot the fitted line
    ax.plot(z, fit_line, linestyle='--', label=f'Fit {z_label}: y = {m:.2f}x + {c:.2f}')

     # Print the fitted m-parameter multiplied by 10 with unit "Gauss/cm"
    print(f'{z_label} vs {B_label} slope (m): {m * 10:.2f} Gauss/cm')

    # Find the x value where the fitted line crosses y = 0
    if m != 0:  # Avoid division by zero
        x_intercept = -c / m
        plt.axvline(x=x_intercept, color='gray', linestyle=':', label=f'{z_label} intercept')

# Add a horizontal line at y = 0
plt.axhline(y=0, color='black', linestyle='-')

ax.legend()

ax.set_title(r' MOT coil B-field measurement')
ax.set_xlabel('Position (mm)')
ax.set_ylabel('Magnetic field (G)')
```
